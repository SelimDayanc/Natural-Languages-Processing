```python
import nltk
import re
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from collections import Counter
#Python Counter takes in input a list, tuple, dictionary, string, which are all iterable objects, and it will give you output that will have the count of each element.
#download collections module( list, dict, set, tuple)
```

```python
def filtered_text(text):
    wnl = nltk.stem.WordNetLemmatizer()
    stopwords = nltk.corpus.stopwords.words('english') 
    tokens = text.split()
    return [wnl.lemmatize(word) for word in tokens if word not in stopwords]
```

```python
data = open("eng_news_2016_10K-sentences.txt", encoding="utf8")
text=data.read()
tokens = filtered_text(text)
```
