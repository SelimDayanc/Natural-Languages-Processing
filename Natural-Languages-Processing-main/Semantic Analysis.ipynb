```python
import nltk
import re
import unicodedata
import pandas as pdq
import numpy as npq
from nltk.corpus import stopwords 
from sklearn.decomposition import TruncatedSVD
from nltk.corpus import stopwords
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = open(r'dünya_tarihi_corpus.txt', 'r', encoding='utf-8')
text=bodytext.read()
```

```python
def filtered_text(text):
    wnl = nltk.stem.WordNetLemmatizer()
    stopwords = nltk.corpus.stopwords.words('turkish') 
    words = re.sub(r'[^\w\s]', '', text).split()
    return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

```python
words = filtered_text(text)
words
```

```python
from nltk.probability import FreqDist
fdist=FreqDist(words)
fdist.most_common(20)
```

```python
bigrams=nltk.ngrams(words,2)
bigramsFrequency=Counter(bigrams)
valuesOfBigrams=list(bigramsFrequency.values())
bigramlist=list(bigramsFrequency)

for x in range(0,len(bigramlist)):
    if(valuesOfBigrams[x]>=5):
        print(bigramlist[x],valuesOfBigrams[x],"times")
```

```python
trigrams = nltk.ngrams(words,3)
trigramsFrequency = Counter(trigrams)
valuesofTrigrams = list(trigramsFrequency.values())
trigramlist = list(trigramsFrequency)

for x in range(0, len(trigramlist)):
    if(valuesofTrigrams[x] >= 5):
        print(trigramlist[x], valuesofTrigrams[x], " times")
```

```python
bodytext = open(r'dünya_tarihi_corpus.txt', 'r', encoding='utf-8')
body = bodytext.readlines()
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(body)
bag_of_words.todense()
```

```python
svd = TruncatedSVD(n_components=1)
lsa = svd.fit_transform(bag_of_words)
topic_encoded_df = pd.DataFrame(lsa, columns=["topic_1"])
topic_encoded_df["body"] = body
print(topic_encoded_df[["body", "topic_1"]])
```

```python
dictionary = vectorizer.get_feature_names()
print(dictionary)
```

```python
encoding_matrix = pd.DataFrame(svd.components_, index=['topic_1'], columns=dictionary).T
print(encoding_matrix)
```
